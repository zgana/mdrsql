```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

```python
N = 10
n = np.arange(N+1)
p = .5

x = n
y = stats.binom(N, p).pmf(n)
fig, ax = plt.subplots()
#ax.bar(x, y)
ax.plot(x, y, '.')
ax.grid(alpha=.2)
```
